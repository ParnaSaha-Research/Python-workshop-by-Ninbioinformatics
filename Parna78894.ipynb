```python
# Consider you have a DNA stock concentration of 500 ng/ul
# You need to dilute it to a specific volume for a PCR reaction
# Find the final concentration
# Formula: Final_Conc = (Stock_Conc * Stock_Volume) / Final_Volume
# Solve this using Python

# Step 1: Assign Variables
stock_conc = 500       # in ng/ul
stock_vol = 2          # amount of DNA you pipetted in ul
final_vol = 50         # total volume after adding water/buffer in ul

# Step 2: Create the formula
final_conc = (stock_conc * stock_vol) / final_vol

# Step 3: Calculation (The computer does this internally)
# final_conc = (500 * 2) / 50 = 20

# Step 4: Frame a sentence
print("DNA stock with concentration of", stock_conc, "ng/ul was used.")
print("After adding", stock_vol, "ul into a total volume of", final_vol, "ul,")
print("the final concentration is", final_conc, "ng/ul.")
```

```python
# Consider you are testing the viability of a new seed lot
# You planted a specific number of seeds and counted how many sprouted
# Find the Germination Percentage
# Formula: (Sprouted / Total) * 100
# Solve this using Python

# Step 1: Assign Variables
total_seeds = 150
sprouted_seeds = 126

# Step 2: Create the formula
germination_rate = (sprouted_seeds / total_seeds) * 100

# Step 3: Calculation (The print function shows the result)
print(germination_rate)

# Step 4: Frame a sentence
print("In a trial of", total_seeds, "seeds,", sprouted_seeds, "successfully sprouted.")
print("The final germination rate for this seed lot is", germination_rate, "%.")
```

```python
# Variables from your Notebook
gene_length = 1234
gc_content = 67.38

# 1. Calculate AT Percentage
# Since Total = 100%, AT % is 100 minus GC %
at_content_percent = 100 - gc_content

# 2. Calculate actual number of AT base pairs
# (AT % / 100) * total length
at_base_pairs = (at_content_percent / 100) * gene_length

# 3. Print the results clearly
print(f"AT Content: {at_content_percent:.2f}%")
print(f"Estimated AT Base Pairs: {at_base_pairs:.0f} bp")
```

In Python, there are two primary ways to remove data from a dictionary. Choosing between them depends on whether there is a need to save the deleted data or just discard it.

1. Using .pop() (The "Save & Delete" Method)
This removes the key and returns the value. It is the safest and most common method in Bio-IT because it allows to move "failed" data into a separate variable before deleting it from the main set.

```python
# Removes 'QC_Pass' and saves its value (True/False) into 'status'
status = sample_metadata.pop("QC_Pass")

print(f"Removed status: {status}")
print(sample_metadata)
```

2. Using del (The "Permanent Delete" Method)
This is a standard Python keyword that simply deletes the entry. If the key doesn't exist, it will throw an error.
