Consider you are preparing buffer for Checking Enzyme activity at specific pH,
Consider a following Ranges of pH:

* pH 0.00 to 3.00 = Highly Acidic
* pH 3.00 to 6.00 = Acidic
* pH 6.00 to 6.50 = Slightly Acidic
* pH 6.60 to 7.40 = Neutral
* pH 7.50 to 8.00 = Slightly Alkaline
* pH 8.00 to 11.00 = Alkaline
* pH 11.00 to 14.00 = Highly Alkaline

Now, You are creating various buffers for the experiment, check Comments for random pH of buffer, using if, elif and else in Python!

```python
buffer_pH = 7.50 #Whatever value can be written

#Concept of Tuples is used, because ranges are fixed!

highly_acidic_range = (0.00,3.00)
acidic_range = (3.00,6.00)
slightly_acidic_range = (6.00,6.50)
neutral_range = (6.50,7.50)
slightly_alkaline_range = (7.50,8.00)
alkaline_range = (8.00,11.00)
highly_alkaline_range = (11.00,14.00)

if buffer_pH >= highly_acidic_range[0] and buffer_pH <= highly_acidic_range[1]:
  print("Highly Acidic")
elif buffer_pH >= acidic_range[0] and buffer_pH <= acidic_range[1]:
  print("Acidic")
elif buffer_pH >= slightly_acidic_range[0] and buffer_pH <= slightly_acidic_range[1]:
  print("Slightly Acidic")
elif buffer_pH >= neutral_range[0] and buffer_pH <= neutral_range[1]:
  print("Neutral")
elif buffer_pH >= slightly_alkaline_range[0] and buffer_pH <= slightly_alkaline_range[1]:
  print("Slightly Alkaline")
elif buffer_pH >= alkaline_range[0] and buffer_pH <= alkaline_range[1]:
  print("Alkaline")
elif buffer_pH >= highly_alkaline_range[0] and buffer_pH <= highly_alkaline_range[1]:
  print("Highly Alkaline")
else:
  print("Invalid pH Value")
print("pH Range Checked!")
```

***Task 10.1.1***

You have Provided with Bacterial Suspension of containing 100 cells, and Incubation Process must be stopped at 1600 cells. find out at which cycle of cell division, process must be stopped?


```python
# Initial population and maximum capacity
current_population = 100
max_population = 1600
division_cycles = 0

# Decoration
print("\n-----Starting the Process-----")

print(f"\nStarting population: {current_population}")

# Loop continues as long as the current population is less than the max
while current_population < max_population:
    division_cycles += 1
    # this += will add RHS value (this case 1) to LHS Value after each cycle
    # Population doubles each cycle (using the *= operator for shorthand multiplication; This will multiply LHS by RHS in each step)
    current_population *= 2

    print(f"Cycle {division_cycles}: Population after cycle {division_cycles} : {current_population}")
print(f"\nPopulation reached maximum capacity after {division_cycles} cycles.")
```

# **Defining Functions (Part 1) - Reusable Code**

Functions are named blocks of reusable code that perform a specific task. They are the single most important concept for writing clean, efficient, and scalable code.

***Task 12.1.1***

Calculate the GC Content of any DNA Sequence

```python
#step 1: Defining the Function/Program
def calculate_gc_content(dna_sequence):

    # def - I'm Defining the Program
    # function name - name of program
    # parameter- Variable/function on which we have to work in this program
    # syntax for def function
    # def (space) function_name(parameter_name):
    """
    Calculates the percentage of Guanine (G) and Cytosine (C)
    bases in a given DNA sequence.
    """

    # 1. Ensure the sequence is uppercase for robust counting
    sequence_upper = dna_sequence.upper()

    # 2. Count G and C bases
    g_count = sequence_upper.count('G')
    c_count = sequence_upper.count('C')
    total_gc = g_count + c_count

    # 3. Get the total length
    total_length = len(sequence_upper)


    # Prevent DivisionByZeroError if sequence is empty
    if total_length == 0:
       return 0.0

    # 4. Calculate the percentage and return the result
    gc_percentage = (total_gc / total_length) * 100

    return gc_percentage

# ----------------------------------------------------
print("Disclaimer:- Those who are siuggesting Long Sequences they will be responsible if System gets Hang!")
dna_strand = input("Enter a DNA sequence: ")
gc_content = calculate_gc_content(dna_strand)
print(f"GC Content: {gc_content:.2f}%")
```

# **Defining Functions (Part 2) - Flexible Arguments**

Functions can be made more flexible using default, positional, and keyword arguments, allowing users to control how the function behaves.

Theory:

Function Arguments

Positional Arguments: Arguments passed in the order defined e.g., function_name(value1, value2).

Keyword Arguments: Arguments specified by name e.g., function_name(name='x', count=5).

Default Arguments: Parameters given a default value in the function definition e.g., def function_name(x, y=10):.

If the user doesn't provide the value for y, it defaults to 10.

```python
def calculate_gc_content_flexible(dna_sequence, as_percentage=True):
    # Syntax for def function
    # def (space) function_name(Parameter1, parameter2):
    """
    Calculates the GC content.
    The 'as_percentage' argument is a default argument, defaulting to True.
    """
    sequence_upper = dna_sequence.upper()
    total_gc = sequence_upper.count('G') + sequence_upper.count('C')
    total_length = len(sequence_upper)

    if total_length == 0:
        return 0.0

    gc_fraction = total_gc / total_length

    # Check the boolean argument to decide the return format
    if as_percentage:
        return gc_fraction * 100
    else:
        return gc_fraction

# ----------------------------------------------------
print("Disclaimer: Participant suggesting long seq will be responsible for Crashing the system!")
test_sequence = input("Enter DNA Sequence: ")

# 1. Call 1: Positional argument only (Uses default: as_percentage=True)
gc_percent = calculate_gc_content_flexible(test_sequence)
print(f"1. Default (Percentage): {gc_percent:.2f}%")

# 2. Call 2: Keyword argument (Overrides default: as_percentage=False)
gc_fraction = calculate_gc_content_flexible(test_sequence, as_percentage=False)
print(f"2. Fraction (Decimal): {gc_fraction:.4f}")

# 3. Call 3: Positional argument used to override the default (same as #2)
gc_fraction_pos = calculate_gc_content_flexible(test_sequence, False)
print(f"3. Fraction (Positional): {gc_fraction_pos:.4f}")
```

# **Revision and Mini Project-Day 14: Protein Length Classifier**



in this mini project, We need to input protein sequences, and as per the criteria, it will classify the sequences in short, medium and long categories, based on length of sequence.

```python
def classify_proteins(length_list):
    """
    Takes a list of protein lengths (integers) and classifies them
    into size categories, returning the counts in a dictionary.
    """
    # Initialize the dictionary to store the counts (keys are size categories)
    classification_counts = {
        "Short": 0,    # Length < 200
        "Medium": 0,   # 200 >= Length <= 550
        "Long": 0      # Length > 500
        #LHS is Key; RHS is Value, in Dictionary, things are always in key:value pair
    }

    # Iterate through the list of lengths
    for length in length_list:

        # Use conditional logic to assign classification
        if length < 200:
            classification_counts["Short"] += 1
            #concept of addition of key value pair in dictionary!

        elif length <= 550: # We already know it's <= 550 from the 'if' failing
            classification_counts["Medium"] += 1

        else: # Must be > 550
            classification_counts["Long"] += 1

    return classification_counts

# --- Test Data and Execution ---
p1 = input("enter protein seq1:")
lenp1 = len(p1)
p2 = input("enter protein seq2:")
lenp2 = len(p2)
p3 = input("enter protein seq3:")
lenp3 = len(p3)
p4 = input("enter protein seq4:")
lenp4 = len(p4)
p5 = input("enter protein seq5:")
lenp5 = len(p5)
p6 = input("enter protein seq6:")
lenp6 = len(p6)
protein_lengths = [lenp1, lenp2, lenp3, lenp4, lenp5, lenp6]

# Call the function
result_dict = classify_proteins(protein_lengths)

print("--- Protein Classification Results ---")
print(f"Input Lengths: {protein_lengths}")
print("Classification Counts:")

# Loop through the resulting dictionary for clean output
for size, count in result_dict.items():
    print(f"  {size}: {count}")
```
